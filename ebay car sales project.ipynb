## Autos cleaning and Exploration

```python
autos.describe(include ='all')
```

when looking at the descriptive statistics for both numerical and string based columns we can identify a few outliers that need further attention. For example, the numerical column "num_photos" has the value 0 for almost all statistic entries. It could be beneficial to drop this cell as it does not provide any useful information about the dataset. The following will be a list of things that need investigation:
- The date columns need to be converted to datetime objects
- the odometer and price columns needs to be cleaned and converted to int type

First, we will convert the "Price" and "Odometer" column to numeric type.


```python
autos.head()
```

```python
autos["price"] = autos["price"].str.replace("$","").str.replace(",","")
autos["price"] = autos["price"].astype(int)

autos["odometer"] = autos["odometer"].str.replace(",","").str.replace("km","")
autos["odometer"] = autos["odometer"].astype(int)
```

```python
autos.rename({"odometer":"odometer_km"}, axis = 1, inplace = True)
```

```python
autos
```

```python
#autos["price"].unique().shape
#autos["price"].describe()
autos_price_counts = autos["price"].value_counts()
autos_price_counts.sort_index(ascending=True).head(50)

```

```python
# for the higher end of the prices spectrum, cut off anything 10000000 and over 
# for the lower end of the price spectrum, remove anything 100 or under
```

using the value_count and describe method on the price row, we discover that several thousand cars had non-intuitive prices. For example, some cars were going for under 100 while other were exceediing 10 million. By removing these entries, we can control for more realistic values

```python
autos_odometer_counts = autos["odometer_km"].value_counts()
autos_odometer_counts.sort_index(ascending=True).head(50)
```

# Exploring price by Brand

```python
autos.columns
```

```python
autos["brand"].value_counts(normalize=True)

len(autos["brand"].unique())
```

German cars represent over 50% of the cars listed on Ebay. Volkswagon being the most common vehicle with rougly 20% of the listings. 

```python
brand_counts = autos["brand"].value_counts(normalize=True)
top_brands = brand_counts[brand_counts > 0.04].index
print(top_brands)
```

The brand column consists of 40 unique car brands, all of which have a different listing share on Ebay. Using sort values it becomes clear that german cars are of the most listed vehicle brands on ebay, accounting for over 50% of the ads. Several car brands in the bottom of the list have very minor share of ebay listings.

For our aggregation purposes we will only keep the any car brands that account for over 4% of vehicle ads on the website. The brands are: 'Volkswagon', 'bmw', 'opel', merceded, audi, ford, and renault.

```python
brand_mean_prices = {}

for brand in top_brands:
    brand_only = autos[autos["brand"] == brand]
    mean_price = brand_only["price"].mean()
    brand_mean_prices[brand] = int(mean_price)

brand_mean_prices
```

after analyzing the mean prices of the top car brands on ebay we see that renault's have the lowest prices of the bunch, while audi had the highest average prices on the ads. Volkswagon is in the middle.

# Checking Mileage by Brand

```python
brand_mean_mileage = {}

for brand in top_brands: 
    specific_brand = autos[autos["brand"] == brand]
    mean_mileage = specific_brand["odometer_km"].mean()
    brand_mean_mileage[brand] = int(mean_mileage)
    
brand_mean_mileage
    
```

To find out if there's any visible link between a veehicle's mileage and its price, we have derived both the top_brand's mean price and mean mileage. Furthermore, we can convert both dictionaries to series and then a datatframe. 

```python
bmp_series = pd.Series(brand_mean_prices) # Using the series constructor we convert the brand price dictionary to a sereis
print(bmp_series)
bmm_series = pd.Series(brand_mean_mileage)
print(bmm_series)
```

Using the first series object we will create a datarame (df) and use the top_brands as the index. After that, we will add another column containing the mean mileage values.

```python
brand_info = pd.DataFrame(bmp_series, columns=['mean_price']) 
brand_info
```

```python
brand_info["mean_mileage"] = bmm_series #Adding mean_mileage series as a column in brand info
brand_info
```

By looking at the brand_info dataframe above we can deduce. bmw's have the highest average mileage and the second highest lsiting prices. Renaults' have the lowest pricing with second lowest avg mileage.
