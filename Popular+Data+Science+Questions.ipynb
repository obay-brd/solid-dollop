```python
import pandas as pd
```

```python
stack = pd.read_csv("2019_questions.csv")
```

```python
stack
```

```python
for col in stack.columns:
    counts = stack[col].value_counts()
    print(counts)
```

```python
missing_vals = stack.isna().sum()
print(missing_vals)
print("\n")
print(stack.info())
```

Only the "FavoriteCount" column contains missing values, to the degree of 7432 na values. The missing values cannot be cleaned as it would remove a massive chunk of valuable information from other columns as well. The "FavoriteCount" column refers to the column contains information on the number of times that a question post was favourited. Thus by that intuition, the rows containing missing values were for posts that weren't deemed question from other users. 

```python
stack["FavoriteCount"] = stack["FavoriteCount"].fillna(0).astype(int)
stack["CreationDate"] = pd.to_datetime(stack["CreationDate"])
stack["AnswerCount"] = stack["AnswerCount"].astype(int)
stack.info()
```

```python
stack["Tags"] = stack['Tags'].str.replace('^<|>$', '').str.split('><')
stack["Tags"]
```

```python
all_tags = [] # Create a list that contains the names of all tags in the colum
for tag in stack["Tags"]:
    all_tags += tag
all_tags
```

```python
tag_counts = pd.Series(all_tags).value_counts().sort_values(ascending=False)
tag_counts # Displays the number of times each tag was used in "Tags" column.
top_tag_counts = tag_counts.nlargest(10)
top_tag_counts
```

The top 5 most common tags for questions in the stack dataset are:
1. machine-learning
2. python
3. deep-learning
4. neural-network
5. keras


## Number of Times Each tag was Viewed

```python
stack
```

```python
tag_view_count = {}        

for index, row in stack.iterrows():
    for tag in row['Tags']:
        if tag not in tag_view_count:
            tag_view_count[tag] = row['ViewCount']
        else:
            tag_view_count[tag] += row['ViewCount']
# Creates a dictionary and uses for loops to iterate the number of views for each tag

tag_view_count = pd.Series(tag_view_count).sort_values(ascending=False)
# Convert the dictionary into a series object 
top_tag_view_count = tag_view_count.nlargest(10)
```

Above we see the tags that bring in the most amount of views per post. The Following are the most popular tags in view_counts:
1. Python
2. Machine-Learning
3. Keras
4. Deep-Learning
5. Pandas

## Deep-Learning Analysis

```python
all_questions = pd.read_csv("all_questions.csv")
```

```python
#all_questions["Tags"] = all_questions['Tags'].str.replace('^<|>$', '').str.split('><')
all_questions
```

```python
all_questions.info()
# We need to change the CreationDate to datetime object in order to aggregate
all_questions["CreationDate"] = pd.to_datetime(all_questions["CreationDate"])
```

```python
all_questions["Tags"] = all_questions['Tags'].str.replace('^<|>$', '').str.split('><')
```

```python
# Use the apply method to find the number of times "deep-learning" was used in each of the posts. 
# Convert the column to int in order to tally up the total number of counts

all_questions["deep_learning_counts"] = all_questions["Tags"].apply(lambda x: "deep-learning" in x).astype(int)
all_questions
```

```python
# We will group the deep-learning (DL) counts by month for each of the years.

monthly_DL_counts = all_questions.groupby(all_questions["CreationDate"].dt.strftime('%Y-%m')).agg({'deep_learning_counts':'sum'})
monthly_DL_counts
```

```python
monthly_questions = all_questions.groupby(all_questions["CreationDate"].dt.strftime('%Y-%m')).size()
monthly_questions
```

```python
# We now convert the monthly questions series into a dataframe 
monthly = monthly_questions.to_frame(name="questions")
monthly

# now we add the deep-learning related questions to this dataframe
monthly["Deep-Learning questions"] = monthly_DL_counts

# Create a column that shows the share of deep-learning questions as a share of total questions in each month
monthly["Share of DL"] = (monthly["Deep-Learning questions"] / monthly["questions"])*100
monthly = monthly.reset_index()
monthly["CreationDate"] = pd.to_datetime(monthly["CreationDate"])
```

```python
plt.plot(monthly["CreationDate"], monthly["Share of DL"])
plt.title("Trends in Deep-Learning Related Questions (DDSE)")
plt.xlabel("Year")
plt.ylabel("DL Questions as Share of Total")
```

Key Observations:
- Total number of questions on is rising with every year
- The total number of deep-learning related questions makes up a small portion (less than 25%) of total questions asked each given year
- There is a noticeable increase in interest for deep-learning as see through the time series. The interest peaked in mid-2018 and is seemingly declining since
