```python
import pandas as pd
sw = pd.read_csv("star_wars.csv", encoding="ISO-8859-1")
print(sw.head(10))
print("\n")
print(sw.columns)
```

```python
yes_no = {
    "Yes": True, 
    "No": False
}
```

```python
sw['Have you seen any of the 6 films in the Star Wars franchise?'] = sw['Have you seen any of the 6 films in the Star Wars franchise?'].map(yes_no)
sw['Have you seen any of the 6 films in the Star Wars franchise?'].value_counts()
```

```python
sw['Do you consider yourself to be a fan of the Star Wars film franchise?'] = sw['Do you consider yourself to be a fan of the Star Wars film franchise?'].map(yes_no)
sw['Do you consider yourself to be a fan of the Star Wars film franchise?'].value_counts()
```

```python
sw.head()
```

# Cleaning and Mapping Checkbox Columns

```python
sw_1 = sw.copy()
survey_cols = ["Which of the following Star Wars films have you seen? Please select all that apply.", "Unnamed: 4", "Unnamed: 5", "Unnamed: 6", "Unnamed: 7", "Unnamed: 8"]

sw_1[survey_cols] = pd.notna(sw_1[survey_cols]) # Uses boolean masking to remove convert values in the column from NaN to False and True otherwise

sw_1
```

```python
sw_1 = sw_1.rename(columns={"Which of the following Star Wars films have you seen? Please select all that apply.":"seen_1", "Unnamed: 4":"seen_2", "Unnamed: 5": "seen_3", "Unnamed: 6": "seen_4", "Unnamed: 7":"seen_5", "Unnamed: 8":"seen_6"})

sw_1
```

# Cleaning the Ranking Columns

```python
sw_1[sw_1.columns[9:15]] = sw_1[sw_1.columns[9:15]].astype(float)

sw_1 = sw_1.rename(columns={"Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.":"ranking_1", "Unnamed: 10":"ranking_2", "Unnamed: 11":"ranking_3", "Unnamed: 12":"ranking_4", "Unnamed: 13":"ranking_5", "Unnamed: 14":"ranking_6"})
```

```python
sw_rankings = sw_1.iloc[:,9:15]
sw_rankings
```

# Finding Highest-Ranked Movie

```python
sw_ranking_means = sw_rankings.mean()
sw_ranking_means
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

sw_ranking_means.plot(kind='bar', legend=False)
plt.xlabel("Ranking_movie")
plt.ylabel("Survey ranking")
plt.title("Average Rankings for Each Stars Wars Movie")
plt.show()
                    
```

So far the survey columns have been cleaned and converted to boolean masks with values being either (True) or (False).
In addition the ranking columns which asked respondants to rank each of the six Star Wars movies have been renamed to for ease of analysis. The renamed cols take the form "ranking_star was movie number". After that, pd.dataframe.mean() was used to calculate the mean ranking of each movie (lower number = higher ranking). The means of movie were plotted into the bar graph above. Results showed the following:
- The fifth movie in the franchise was ranked the highest by most survey participants
- The third movie was ranked lowest. 

# Finding The Most Viewed Movie

```python
seen_cols = sw_1.iloc[:,3:9]
seen_cols_tally = seen_cols.sum()
seen_cols_tally
```

```python
seen_cols_tally.plot(kind='bar', legend=False)
plt.title("Most Viewed Stars Wars Movie")
plt.show()
```

The results tell us that in addition to being the fan favourite title, the fifth movie in the Star Wars franchise is also the most viewed of them all. The third movie was the least viewed of the bunch. In comparison to the ranking graph, the results in this graph are proportionately correlated with views of each movie. That is, movies with higher viewership tend to be ranked higher amongst the movies in the Star Wars series.  

# Exploring the Data by Gender

```python
males = sw_1[sw_1["Gender"] == "Male"]
females = sw_1[sw_1["Gender"] == "Female"]
```

```python
# Creating dataframes that contain only information about the ranking for each gender.
sw_ranking_males = males.iloc[:,9:15]
sw_ranking_females = females.iloc[:,9:15]
# Calculate the means using pd.df.mean() for each gender df
sw_ranking_means_m = sw_ranking_males.mean()
sw_ranking_means_f = sw_ranking_females.mean()

print(sw_ranking_means_m)
print("\n")
print("\n")
print(sw_ranking_means_f)
```

```python
plt.subplot(1,2,1)
sw_ranking_means_m.plot(kind='bar', legend=False)
plt.title("Star Wars Movie Rankings for Males")

plt.subplot(1,2,2)
sw_ranking_means_f.plot(kind='bar',legend=False)
plt.title("Star Wars Movie Rankings for Females")
plt.subplots_adjust(left=0.001, right=1.5, bottom=0.001,top=0.9)
plt.show()
```

Both Gender groups ranked the 5th movie as the best out of the bunch. Also, the third movie seemed to rank lowest for both groups. The results differ in the scale of rankigs for first, second, and fourth movie. Although the position within the ranks was consistent in both groups, males' mean ranking for the first two movies was higher than that of females. At the same time, females' mean ranking of the fourth movie was higher than that of the males group.  

```python
# Creating dataframes that select only the viewership coloumns for each gender
seen_cols_males = males.iloc[:,3:9]
seen_cols_females = females.iloc[:,3:9]
# Tallying the number of view for each dataframe using sum() command
seen_tally_males = seen_cols_males.sum()
seen_tally_females = seen_cols_females.sum()

print(seen_tally_males)
print("\n")
print("\n")
print(seen_tally_females)
```

```python
plt.subplot(1,2,1)
seen_tally_males.plot(kind='bar', legend=False)
plt.title("Views per Movie for Males")

plt.subplot(1,2,2)
seen_tally_females.plot(kind='bar',legend=False)
plt.title("Views per Movie for Females")
plt.subplots_adjust(left=0.001, right=1.5, bottom=0.001,top=0.9)
plt.show()
```

Viewerships per movie is consistent among both groups with the 5th movie boasting the highest veiwer count and the 3rd being the lowest.

# Star Wars Fans vs. Casual Watchers

```python
sw_fans = sw_1[sw_1["Do you consider yourself to be a fan of the Star Wars film franchise?"] == True]
sw_casual = sw_1[sw_1["Do you consider yourself to be a fan of the Star Wars film franchise?"] == False]
```

```python
# Creating dataframes that contain only information about the ranking for each group.
sw_ranking_fans = sw_fans.iloc[:,9:15]
sw_ranking_casual = sw_casual.iloc[:,9:15]
# Calculate the means using pd.df.mean() for each group df
sw_ranking_means_fans = sw_ranking_fans.mean()
sw_ranking_means_casual = sw_ranking_casual.mean()

print(sw_ranking_means_fans)
print("\n")
print("\n")
print(sw_ranking_means_casual)
```

```python
plt.subplot(1,2,1)
sw_ranking_means_fans.plot(kind='bar', legend=False)
plt.title("Star Wars Movie Rankings for Franchise Fans")

plt.subplot(1,2,2)
sw_ranking_means_casual.plot(kind='bar',legend=False)
plt.title("Star Wars Movie Rankings for Casual Watchers")
plt.subplots_adjust(left=0.001, right=1.5, bottom=0.001,top=0.9)
plt.show()
```

Fans of the Star wars movie ranked the first three movies much lower than than the casual watcher group did. Also, the last three movies were ranked higher by the franchise fans. 

```python
# Creating dataframes that select only the viewership coloumns for each group
seen_cols_fans = sw_fans.iloc[:,3:9]
seen_cols_casual = sw_casual.iloc[:,3:9]
# Tallying the number of view for each dataframe using sum() command
seen_tally_fans = seen_cols_fans.sum()
seen_tally_casual = seen_cols_casual.sum()

print(seen_tally_fans)
print("\n")
print("\n")
print(seen_tally_casual)
```

```python
plt.subplot(1,2,1)
seen_tally_fans.plot(kind='bar', legend=False)
plt.title("Views per Movie for Star Wars Fans")

plt.subplot(1,2,2)
seen_tally_casual.plot(kind='bar',legend=False)
plt.title("Views per Movie for Casual Watchers")
plt.subplots_adjust(left=0.001, right=1.5, bottom=0.001,top=0.9)
plt.show()
```

Viewership across all six movies was almost twice as high for the franchise fans than it was for the casuals. The second to fourth movie in particular were viewed to a much greater extent by the fans than the casuals.
